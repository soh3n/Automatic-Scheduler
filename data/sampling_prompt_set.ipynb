```python
import json
import random
```

```python
real_email_path = "PAIRED_real_emails.json"
pseudo_email_path = "PAIRED_gpt-4o-mini_pseudo_emails.json"
spam_email_path = "PAIRED_spam_emails.json"
```

```python
def get_json(path):
    with open(path, "r") as f:
        data = json.load(f)
    return data
```

```python
real_emails = get_json(real_email_path)
pseudo_emails = get_json(pseudo_email_path)
spam_emails = get_json(spam_email_path)
```

```python
def get_split(combined_list, ratio=0.6):
    # Shuffle the combined list
    random.shuffle(combined_list)

    # Split the combined list in a 6:4 ratio
    split_index = int(len(combined_list) * ratio)
    train_sample = combined_list[:split_index]
    test_sample = combined_list[split_index:]
    return train_sample, test_sample
```

```python
train_real, test_real =  get_split(real_emails)
train_pseudo, test_pseudo =  get_split(pseudo_emails)
train_spam, test_spam =  get_split(spam_emails)
```

```python
train_real + train_pseudo + train_spam
```

```python
train_all =  train_real + train_pseudo + train_spam
test_all = test_real + test_pseudo + test_spam
random.shuffle(train_all)
random.shuffle(test_all)
```

```python
with open("PAIRED_train.json", "w") as output_file:
    json.dump(train_all, output_file, indent=4)
```

```python
with open("PAIRED_test.json", "w") as output_file:
    json.dump(test_all, output_file, indent=4)
```
